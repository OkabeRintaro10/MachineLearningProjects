```python
import csv
a = []
print("\nGiven Training Data Set:\n")
with open('/Users/futuregadgetlab/Developer/MLProjects/datasets/test.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    for r in reader:
        a.append(r)
        print(r)
s_hypothesis = a[1][:-1]
g_hypothesis = [['?' for i in range(len(s_hypothesis))] for j in
range(len(s_hypothesis))]
for i in a:
    if i[-1] == "yes":
        for j in range(len(s_hypothesis)):
            if i[j] != s_hypothesis[j]:
                s_hypothesis[j] = '?'
                g_hypothesis[j][j] = '?'
    elif i[-1] == "no":
        for j in range(len(s_hypothesis)):
            if i[j] != s_hypothesis[j]:
                g_hypothesis[j][j] = s_hypothesis[j]
            else:
                g_hypothesis[j][j] = '?'
            
    print("\nStep {0}:\n".format(a.index(i)+1))
    print(s_hypothesis, end='\n\n')
    for c in range(len(g_hypothesis)): print(g_hypothesis[c])
    print("")
general_hypothesis = []
for i in g_hypothesis:
    for j in i:
        if j != '?':
            general_hypothesis.append(i)
            break

print("Final Specific Hypothesis: {0}".format(s_hypothesis))
print("Final General Hypothesis: {0}".format(general_hypothesis))
```
